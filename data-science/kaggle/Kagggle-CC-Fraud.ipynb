```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_df = pd.read_csv('creditcard.csv')
```

```python
my_df
```

```python
my_df['Class'].describe()
```

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x='Class', hue='Class', data=my_df) # palette="pastel")
ax.get_yaxis().set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))  #format y in thousands
for i in ax.containers:
    ax.bar_label(i,fmt='{:,.0f}',)
```
